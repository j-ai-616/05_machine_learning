# 회귀 평가

### 회귀 평가지표 요약

| 평가 지표 | 설명 | LaTeX 수식 |
| :--- | :--- | :--- |
| **평균 제곱 오차 (MSE)** | 예측 값과 실제 값의 차이를 제곱하여 평균을 낸 값입니다. | $$\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$ |
| **평균 절대 오차 (MAE)** | 예측 값과 실제 값의 차이의 절대값을 평균한 값입니다. | $$\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} \lvert y_i - \hat{y}_i \rvert$$ |
| **루트 평균 제곱 오차 (RMSE)** | MSE의 제곱근으로, 실제 오차와 단위가 같습니다. 큰 오차에 더 가중치를 둡니다. | $$\text{RMSE} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}$$ |
| **평균 제곱 로그 오차 (MSLE)** | 실제 값과 예측 값의 로그 차이를 제곱하여 평균을 낸 값입니다. | $$\text{MSLE} = \frac{1}{n} \sum_{i=1}^{n} (\log(y_i + 1) - \log(\hat{y}_i + 1))^2$$ |
| **루트 평균 제곱 로그 오차 (RMSLE)** | MSLE의 제곱근으로, 로그 스케일에서 예측 정확도를 측정합니다. | $$\text{RMSLE} = \sqrt{\text{MSLE}}$$ |
| **결정 계수 ($R^2$)** | 모델이 데이터의 변동성을 얼마나 설명하는지 나타냅니다. (1에 가까울수록 우수) | $$R^2 = 1 - \frac{\sum (y_i - \hat{y}_i)^2}{\sum (y_i - \bar{y})^2}$$ |
| **조정된 결정 계수 (Adj. $R^2$)** | 독립 변수의 개수를 고려하여 $R^2$를 보정한 값입니다. | $$1 - \frac{(1-R^2)(n-1)}{n-p-1}$$ |
| **설명 분산 점수** | 예측이 실제 데이터의 분산을 얼마나 잘 설명하는지 나타냅니다. | $$1 - \frac{\text{Var}(y - \hat{y})}{\text{Var}(y)}$$ |
| **평균 절대 백분율 오차 (MAPE)** | 오차를 실제 값으로 나눈 백분율의 절대값 평균입니다. | $$\frac{1}{n} \sum \lvert \frac{y_i - \hat{y}_i}{y_i} \rvert \times 100$$ |
| **최대 절대 오차 (Max Error)** | 발생한 오차 중 가장 큰 절대값을 측정합니다. | $$\max(\lvert y_i - \hat{y}_i \rvert)$$ |
| **Huber Loss** | MSE와 MAE의 장점을 결합하여 이상치에 강건하게 설계된 손실 함수입니다. | $$L_\delta = \begin{cases} \frac{1}{2}(e_i)^2 & \lvert e_i \rvert \le \delta \\ \delta(\lvert e_i \rvert - \frac{1}{2}\delta) & \text{otherwise} \end{cases}$$ |
| **Log-Cosh Loss** | 오차의 하이퍼볼릭 코사인 로그 값을 사용하여 부드러운 곡선을 가집니다. | $$\sum \log(\cosh(\hat{y}_i - y_i))$$ |
| **대수 평균 제곱 오차 (Log-MSE)** | 로그 변환된 값들의 평균 제곱 오차입니다. | $$\frac{1}{n} \sum (\log(y_i + 1) - \log(\hat{y}_i + 1))^2$$ |
