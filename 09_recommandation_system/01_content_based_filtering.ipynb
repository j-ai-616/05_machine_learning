# 추천 시스템

추천시스템은 사용자가 좋아할 가능성이 높은 아이템(영화, 제품, 음악 등)을 추천하는 시스템이다. 사용자와 아이템의 특성, 과거 상호작용, 유사성 등을 바탕으로 개인화된 추천을 제공한다.

추천 시스템의 주요 유형은 다음과 같다.

1. 콘텐츠 기반 필터링(Content-Based Filtering):
사용자가 과거에 선호한 아이템의 속성과 유사한 아이템을 추천하는 방식이다.
예를 들어 사용자가 액션 영화, SF 영화처럼 특정 장르를 자주 봤다면, 비슷한 장르나 속성을 가진 다른 영화를 추천할 수 있다.

2. 협업 필터링(Collaborative Filtering):
비슷한 사용자들의 행동 패턴이나, 함께 소비된 아이템의 관계를 바탕으로 추천하는 방식이다.
즉, 아이템 자체의 속성보다 사용자-아이템 상호작용 기록을 활용한다.

   2-1. 사용자 기반 협업 필터링:
   비슷한 구매나 평가 패턴을 가진 사용자들을 찾아 추천한다.
   예를 들어 홍길동이 라면, 만두, 김치를 구매했고 신사임당이 라면, 만두를 구매했다면, 두 사용자는 비슷한 소비 패턴을 보인다고 판단할 수 있다. 이 경우 신사임당에게 김치를 추천할 수 있다.

   2-2. 아이템 기반 협업 필터링:
   함께 구매되거나 함께 소비되는 경향이 높은 아이템을 추천한다.
   예를 들어 비빔면과 삼겹살을 같이 구매하는 경우가 많다면, 비빔면을 구매한 사용자에게 삼겹살을 추천할 수 있다.

3. 하이브리드 추천 시스템(Hybrid Recommendation Systems):
콘텐츠 기반 필터링과 협업 필터링을 결합한 방식이다.
협업 필터링은 신규 사용자나 신규 아이템에 대한 정보가 부족할 때 콜드 스타트 문제가 발생할 수 있다. 이를 보완하기 위해 하이브리드 방식은 초기에는 콘텐츠 정보도 활용하고, 이후 상호작용 데이터가 충분히 쌓이면 협업 신호를 함께 반영하여 추천 성능을 높인다.
넷플릭스와 같은 대규모 서비스는 이런 다양한 추천 신호를 복합적으로 활용하는 대표적인 사례로 볼 수 있다.

요즘 추천 시스템은 단순히 한 가지 알고리즘만 쓰기보다, 후보를 빠르게 찾는 단계와 그 후보를 정교하게 재정렬하는 단계를 나누어 여러 모델을 함께 사용하는 경우가 많다.

## 코사인 유사도 cosine similiarity

코사인 유사도(Cosine Similarity)는 두 벡터 간의 유사도를 측정하는 방법으로, 벡터의 방향을 비교하여 얼마나 유사한지를 계산하는 지표이다.

코사인 유사도의 값은 -1에서 1 사이에 있으며, 1에 가까울수록 두 벡터의 방향이 유사하다는 것을 의미한다. 주로 문서의 유사도를 계산하거나, 추천 시스템 등에서 사용된다.


**벡터란?**
벡터는 방향과 크기를 가진 데이터다.

집에서 학교까지 걸어간다고 할 때, 벡터로는 두 가지 정보를 나타낼 수 있다.
- 방향: 집에서 학교로 가는 길의 방향.
- 크기: 집에서 학교까지의 거리.

예를 들어, (1, 2)와 같은 2차원 좌표를 벡터라고 할 수 있다.

**Vector의 크기와 방향**

벡터의 방향과 크기를 따로 구할 수 있다.

1. **벡터의 크기 (Magnitude):** 벡터의 크기는 벡터의 각 성분을 제곱한 후 더한 값을 루트 씌운 것으로 구할 수 있다. 예를 들어, 2차원 벡터 $\vec{v} = (x, y)$의 크기는 다음과 같다.

   $
   \text{크기} = \sqrt{x^2 + y^2}
   $

   3차원 벡터의 경우, $\vec{v} = (x, y, z)$라면 크기는 다음과 같다.

   $
   \text{크기} = \sqrt{x^2 + y^2 + z^2}
   $

2. **벡터의 방향 (Direction):** 벡터의 방향은 벡터의 각 성분을 벡터의 크기로 나눈 정규화된 벡터로 구할 수 있다. 이를 단위 벡터라고도 한다. 벡터 $\vec{v} = (x, y)$의 경우 방향은 다음과 같이 구한다.

   $
   \text{단위 벡터} = \left(\frac{x}{\text{크기}}, \frac{y}{\text{크기}}\right)
   $

   3차원 벡터라면,

   $
   \text{단위 벡터} = \left(\frac{x}{\text{크기}}, \frac{y}{\text{크기}}, \frac{z}{\text{크기}}\right)
   $


```python
import numpy as np

def vec_magnitude_direction(vec):
    # 크기 magnitude
    mag = np.sqrt(np.sum(vec ** 2))
    # 방향 (단위벡터)
    if mag != 0:
        dir = vec / mag
    else:
        dir = np.zeros_like(vec)
    return mag, dir

vec  = np.array([3, 4])
print(vec_magnitude_direction(vec))

vec  = np.array([3, 4, 5])
print(vec_magnitude_direction(vec))
```

```python
import numpy as np

def vec_magnitude_direction(vec):
    # 크기 magnitude
    mag = np.linalg.norm(vec)
    # 방향 (단위벡터)
    dir = vec / mag
    return mag, dir

vec  = np.array([3, 4])
print(vec_magnitude_direction(vec))

vec  = np.array([3, 4, 5])
print(vec_magnitude_direction(vec))
```

### 두 벡터의 코사인 유사도

두 벡터가 얼마나 비슷한지 측정하는 방법이다. 코사인 유사도는 벡터의 방향을 기준으로 유사도를 측정한다.

**코사인 유사도의 수식**

코사인 유사도는 두 벡터 사이의 코사인을 계산한다. 벡터 $ A $와 $ B $가 주어지면, 코사인 유사도는 다음과 같이 계산된다:

$
Cosine Similarity = \frac{A \cdot B}{||A|| \times ||B||}
$

- $A \cdot B$는 두 벡터의 내적 (dot product)
- $||A||$와 $||B|| $는 각 벡터의 크기 (magnitude)

**코사인 유사도 계산 단계**

1. 벡터의 내적 계산

    두 벡터의 대응하는 원소들을 곱해서 모두 더한다.

    예: 벡터 $ A = (1, 2, 3) $와 $ B = (4, 5, 6) $일 때,

    $
    A \cdot B = (1 \times 4) + (2 \times 5) + (3 \times 6) = 4 + 10 + 18 = 32
    $

2. 벡터의 크기 계산

    각 벡터의 크기는 벡터 원소들의 제곱을 더한 후 제곱근을 취한다.

    예: 벡터 $ A $와 $ B $의 크기

    $||A|| = \sqrt{1^2 + 2^2 + 3^2} = \sqrt{1 + 4 + 9} = \sqrt{14}$
    <br>
    $||B|| = \sqrt{4^2 + 5^2 + 6^2} = \sqrt{16 + 25 + 36} = \sqrt{77}$

3. 코사인 유사도 계산

    내적을 벡터 크기의 곱으로 나눈다.

    $
    \text{Cosine Similarity} = \frac{32}{\sqrt{14} \times \sqrt{77}} = \frac{32}{\sqrt{1078}} = 0.9746318461970763
    $

4. 최종 값

    계산된 값을 통해 두 벡터의 유사도를 측정한다. 값이 1에 가까울수록 두 벡터는 더 유사하고, -1에 가까울수록 더 다르다.



코사인 유사도는 두 벡터가 얼마나 비슷한 방향을 가지고 있는지를 측정한다.

텍스트 분석 등에서 문서 간의 유사도를 계산하는 데 자주 사용된다. 계산 과정은 벡터의 내적과 크기를 구한 후 이를 이용해 최종적으로 유사도를 구하는 것이다.

텍스트 분석등에서는 count를 기반으로 하므로, 모든 vector가 0이상의 값이다. 즉 0~90° 사이의 값이고, 이에 따르는 코사인 유사도는 0 ~ 1 사이의 값이다.

```python
vec_a = np.array([1, 2, 3])
# vec_a = np.array([-1, -2, -3])
vec_b = np.array([4, 5, 6])

def cos_similarity(A, B):
    # 두 벡터의 내적 / 두 벡터 크기 곱
    dot_product = A @ B
    norm_A = np.linalg.norm(A)
    norm_B = np.linalg.norm(B)
    return dot_product / (norm_A * norm_B)

cos_similarity(vec_a, vec_b)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

vec_a = np.array([1, 2, 3]).reshape(1, -1)
vec_b = np.array([4, 5, 6]).reshape(1, -1)
print(vec_a, vec_b)

# batch차원으로 감싸진 vector를 전달해야 함
cosine_similarity(vec_a, vec_b)
```

### 평가 횟수에 대한 가중치가 부여된 평점(Weighted Rating) 적용


$$
가중 평점(Weighted Rating) = (v/(v+m)) * R + (m/(v+m)) * C
$$


| 기호 | 의미                        | 예시      |
| -- | ------------------------- | ------- |
| v  | 이 영화에 평점을 준 사람 수 vote_count         | 예: 30명  |
| R  | 이 영화의 평균 평점 vote_average               | 예: 9.0점 |
| m  | 믿을 만한 평가가 되려면 필요한 최소 평가 수. 임계치 설정 | 예: 100명 |
| C  | 전체 영화의 평균 평점. 전체적으로 평점이 후한 편인지, 박한 편인지 반영. vote_average               | 예: 7.0점 |
